```python
import librosa
import librosa.display
import soundfile as sf
import matplotlib.pyplot as plt
from matplotlib.pyplot import specgram
import numpy as np
%matplotlib inline
```

```python
SCALE='dB'
UID='<uid>'
AUDIO_PATH = f'./data/audio_data/{UID}/audio_message_0.wav'
ORIGINAL_AUDIO_PATH = f'./data/audio_data/{UID}/original.wav'
```

```python
y, sr = librosa.load(AUDIO_PATH, sr=None)
yo, sro = librosa.load(ORIGINAL_AUDIO_PATH, sr=None)
plt.figure(figsize=(12, 4))

print("Preprocessed audio samplerate: ",sr)
print("Original audio samplerate:", sro)

plt.subplot(1, 2, 1) 
# librosa.display.waveshow(y, sr=sr, color='blue')
a = specgram(y, Fs=sr, mode='psd', scale=SCALE)
plt.title('Preprocessed Audio Spectogram')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.colorbar(label='Intensity (dB)')

plt.subplot(1, 2, 2) 
# librosa.display.waveshow(yo, sr=sro, color='red')
a = specgram(yo, Fs=sro, mode='psd', scale=SCALE)
plt.title('Original Audio Spectogram')
plt.xlabel('Time (s)')
plt.colorbar(label='Intensity (dB)')

plt.show()
```
